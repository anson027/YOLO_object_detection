```python
# Backbone of YOLO(You Only Look Once) - CSPDarkNet architecture (Cross Stage Partial Darknet)
# YOLO - a trained model used to detect the objects 
# Architecture - combination of convolutional and pooling layer
# parameters - weights and bias updated in each convolutional and pooling layers



# kdfnvemfv
# kdngiermg
```

```python
import ultralytics
model=ultralytics.YOLO('yolov8n.pt')   # there are many other yolov8 versions like yolov8n,yolov8s,yolov8m,yolov8l,yolov8x
```

```python
import cv2
```

```python
video=cv2.VideoCapture(0)
while True:
  suc,img=video.read()
  img=cv2.flip(img,1)
  img1=cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
  results=model([img])
  for result in results:
    result.save(filename='result.jpg')
    img=cv2.imread('result.jpg')
    cv2.imshow('Image',img)
  if cv2.waitKey(1) & 0XFF==113:
    break
video.release()
cv2.destroyAllWindows()
```
